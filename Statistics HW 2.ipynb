```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy, scipy.stats
import plotly.express as px
```

```python
def conf_interval_1(max_size = 10000, gamma = 0.95): 
    s = [] 
    w = [] 
    for n in range(1, max_size, 20):
        sample = np.random.normal(0, 5, n) 
        l1 = scipy.stats.chi2(n).ppf((1+gamma)/2)
        l2 = scipy.stats.chi2(n).ppf((1-gamma)/2)
        left = np.sum(sample**2)/l1 
        right = np.sum(sample**2)/l2 
        w.append(right - left)
        s.append(n)

    return w[1:], s[1:]
```

```python
w, s = conf_interval_1(max_size = 10000, gamma = 0.95)
fig = px.line(x=s, y=w, labels={'x':'sample size', 'y':'interval width'})
fig.show()
```

```python
def conf_interval_2(max_size = 10000, gamma = 0.95): 
    s = [] 
    w = [] 
    for n in range(1, max_size):
        sample = np.random.normal(0, 5, n) 
        l1 = scipy.stats.norm().ppf((3 + gamma) / 4) 
        l2 = scipy.stats.norm().ppf((3 - gamma) / 4) 
        left = n * (np.mean(sample))**2 / (l1**2) 
        right = n * (np.mean(sample))**2 / (l2**2) 
        w.append(right - left)
        s.append(n)

    return w[1:], s[1:]
```

```python
w, s = conf_interval_2(max_size = 10000, gamma = 0.95)
fig = px.line(x=s, y=w, labels={'x':'sample size', 'y':'interval width'})
fig.show()
```
